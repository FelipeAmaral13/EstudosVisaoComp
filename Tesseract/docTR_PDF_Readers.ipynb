# Analysis of a document

```python
import matplotlib.pyplot as plt
from doctr.models import ocr_predictor
from doctr.io import DocumentFile
from doctr.io import image


model = ocr_predictor(det_arch='db_resnet50', reco_arch='crnn_vgg16_bn', pretrained=True)
```

We need set to True the following attibutes for get words by line and not all words in just line.

```python
model.doc_builder.resolve_lines = True
model.doc_builder.resolve_blocks = True
```

```python
# doc = DocumentFile.from_images("/content/image_name.jpg")
doc = DocumentFile.from_pdf("/content/drive/MyDrive/teste_s.pdf")


out = model(doc)

out.show(doc)
```

```python
json_output = out.export()
dict(list(json_output.items())[0:2])
```

```python
# save
import unidecode
import json

doc_name = 'Teste_leitura.pdf'
path_to_data = '/content/'

doc_json = unidecode.unidecode(doc_name).lower().replace(' - ','_').replace(' ','_').replace('.pdf','.json')
path_to_doc_json = path_to_data + doc_json


with open(path_to_doc_json, 'w') as json_file:
    json.dump(json_output, json_file)

# load
import json
with open(path_to_doc_json) as json_file:
    json_output = json.load(json_file)
```

```python
words_list = list()
confidence_list = list()
xmin_list = list()
xmax_list = list()
ymin_list = list()
ymax_list = list()

for block in json_output['pages'][0]['blocks']:
  for line in block['lines']:
    for word in line['words']:

      # get coordonnates & word
      geometry = word['geometry']
      confidence = word['confidence']
      xmin = geometry[0][0]
      xmax = geometry[1][0]
      ymin = geometry[0][1]
      ymax = geometry[1][1]
      w = word['value']
      c = word['confidence']
      

      # get coordonnates and words lists without head and tail of the page
      # if ymin >= 0.12 and ymax <= 0.71:
      if ymin >= 0. and ymax <= 1.:
        xmin_list.append(xmin)
        xmax_list.append(xmax)
        ymin_list.append(ymin)
        ymax_list.append(ymax)
        words_list.append(w)
        confidence_list.append(c)
```

```python
import pandas as pd
df = pd.DataFrame({'word':words_list,'confidence':confidence_list, 'xmin':xmin_list,'ymin':ymin_list, 'xmax':xmax_list, 'ymax':ymax_list})
df.head(15)
```

# Color Segmetation

O objetivo desse proframa é fazer uma segmentação da imagem através das cores que a mesma possui.

### O que são "Color Spaces"?

O mais comum "color space" é o famigerado padrão RGB (Red, Green e Blue), tecnicamente falando , o padrão RGB descreve uma cor através de uma matriz com três componentes. Cada componente pode ter valores entre 0 e 255, onde [0,0,0] representa a cor preta e [255,255,255] representa a cor branca.

O RGB é considerado um espaço de cores "aditivo" e as cores podem ser imaginadas como produzidas a partir de quantidades brilhantes de luz vermelha, azul e verde em um fundo preto.

Mas o padrão RGB não é o único que pode representar o espaço de cores de um modelo, existem mais quatro. Existem várias representações dos espaços de cores porque o uso pode ser variado, ter propositos diferentes.

No mundo da impressão, o CMYK é útil porque descreve as combinações de cores necessárias para produzir uma cor a partir de um fundo branco. Enquanto a matriz RGB é preta, em CMYK a matriz 0 é branca. Nossas impressoras contêm cartuchos de tinta ciano, magenta, amarelo e preto.

Em certos tipos de campos médicos, as lâminas de vidro montadas com amostras de tecido manchado são digitalizadas e salvas como imagens. Eles podem ser analisados no espaço HED, uma representação das saturações dos tipos de manchas - hematoxilina, eosina e DAB - aplicadas ao tecido original.

HSV e HSL são descrições de matiz, saturação e brilho / luminância, que são particularmente úteis para identificar o contraste nas imagens. Esses espaços de cores são freqüentemente usados em ferramentas de seleção de cores em software e para web design.

Na realidade, a cor é um fenômeno contínuo, o que significa que há um número infinito de cores. Os espaços de cores, no entanto, representam cores através de estruturas discretas (um número fixo de valores inteiros de número inteiro), o que é aceitável, pois o olho humano e a percepção também são limitados. Os espaços de cores são totalmente capazes de representar todas as cores que podemos distinguir.

Agora que entendemos o conceito de espaços de cores, podemos usá-los no OpenCV.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

img = cv2.imread(os.path.join(os.getcwd(),'1.jpg'))
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Plot das imagens Original e em RGB
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, constrained_layout=False, figsize=(16,9))
ax1.imshow(img)
ax1.set_xlabel("Imagem Original", fontsize=14)
ax2.imshow(img_RGB)
ax2.set_xlabel("Imagem RGB", fontsize=14)
plt.show()

```

### Visualizando da imagem no espaço de cores RGB

O HSV é uma boa escolha de espaço de cores para segmentar por cor, mas para ver por que, vamos comparar a imagem nos espaços de cores RGB e HSV visualizando a distribuição de cores de seus pixels. Um gráfico 3D mostra isso muito bem, com cada eixo representando um dos canais no espaço de cores. Se você quiser saber como fazer um gráfico 3D.

```python
#Vizualizacao do Espaco em RGB
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm, colors

r,g,b = cv2.split(img)
fig = plt.figure()
axis = fig.add_subplot(1,1,1, projection="3d")
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1., vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()

```

A partir desse gráfico, você pode ver que as partes alaranjadas da imagem abrangem quase todo o intervalo de valores de vermelho, verde e azul. Como partes do Nemo se estendem por todo o gráfico, segmentar o Nemo no espaço RGB com base em intervalos de valores RGB não seria fácil.

### Visualizando a imagem no espaço de cores HSV

Vimos Nemo no espaço RGB, agora vamos vê-lo no espaço HSV e comparar.

Como mencionado brevemente acima, HSV significa Matiz, Saturação e Valor (ou brilho) e é um espaço de cores cilíndrico. As cores, ou matizes, são modeladas como uma dimensão angular que gira em torno de um eixo vertical central, que representa o canal de valor. Os valores vão de escuro (0 na parte inferior) a claro na parte superior. O terceiro eixo, saturação, define os tons de matiz de menos saturado, no eixo vertical, para o mais saturado mais distante do centro:

```python
#Vizualizacao do Espaco em HSV

hsv_img = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)

h,s,v = cv2.split(hsv_img)
fig = plt.figure()
axis = fig.add_subplot(1,1,1, projection="3d")
axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()

```

No espaço HSV, as laranjas de Nemo são muito mais localizadas e visualmente separáveis. A saturação e o valor das laranjas variam, mas estão localizados principalmente dentro de uma pequena faixa ao longo do eixo da tonalidade. Este é o ponto principal que pode ser aproveitado para segmentação.

### Threshold das cores

É preciso traçar um limiar de separação para podermos segmentar a imagem através das cores.

```python
#Threshold baseado nos niveis de laranja
light_orange = (75, 205, 255)
dark_orange = (105, 205, 255)

mask = cv2.inRange(hsv_img, light_orange, dark_orange)
result = cv2.bitwise_and(img_RGB, img_RGB, mask=mask)
plt.figure(figsize=(16,9))
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()

```

```python
#Segunda máscara para as partes azuis
light_white = (0, 0, 200)
dark_white = (145, 60, 255)

mask_white = cv2.inRange(hsv_img, light_white, dark_white)
result_white = cv2.bitwise_and(img_RGB, img_RGB, mask=mask_white)
plt.figure(figsize=(16,9))
plt.subplot(1, 2, 1)
plt.imshow(mask_white, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result_white)
plt.show()

```

```python
#Adicao das máscaras
final_mask = mask + mask_white
final_result = cv2.bitwise_and(img_RGB, img_RGB, mask=final_mask)
plt.figure(figsize=(16,9))
plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.show()

```

A segmentação foi concluida mas é preciso melhorar a imagem, vamos passar um filtro mediano. 

```python
#Melhorando a imagem

median = cv2.medianBlur(final_result,3)
plt.figure(figsize=(16,9))
plt.imshow(median)
plt.show()
```

### Segmentação aplicada a outras imagens


```python
folder=os.getcwd()

images = []
for filename in os.listdir(folder):
    img = cv2.imread(os.path.join(folder,filename))
    if img is not None:
        images.append(img)
    
    
    
```

```python
def segment(image):
    
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    light_orange = (1, 190, 200)
    dark_orange = (18, 255, 255)

     
    mask = cv2.inRange(hsv_image, light_orange, dark_orange)

    
    light_white = (0, 0, 200)
    dark_white = (145, 60, 255)

    
    mask_white = cv2.inRange(hsv_image, light_white, dark_white)

    
    final_mask = mask + mask_white
    result = cv2.bitwise_and(image, image, mask=final_mask)

    
    blur = cv2.medianBlur(result, 3)
    return blur
```

```python
results = [segment(img) for img in images]
```

```python
for i in range(1, 3):
    plt.figure(figsize=(16,9))
    plt.subplot(2, 2, 1)
    plt.imshow(images[i])
    plt.subplot(1, 2, 1)
    plt.imshow(results[i])
    plt.show()
```
